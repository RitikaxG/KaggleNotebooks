# **Distribution Plot for Numeric Columns**

```python
# Distribution of Age
sns.displot(df['Age'])
```

```python
# Distribution of Age
# Displaying more than one graph at a time
sns.distplot(df['Age'],kde=True,hist=True,rug=True)
```

Maximum people are of the age group 30-70

```python
# Divide the whole dataset into 10 bins
sns.distplot(df['Age'],bins=10)
```

```python
sns.distplot(df["RestingBP"],kde=True,color='red')
```

Most people are having RestingBP from 90 to 180

```python
sns.distplot(df["Cholesterol"],kde=True,color='green')
```

Most people are having cholestrol from 100 to 400

```python
sns.distplot(df["MaxHR"],kde=True,color='pink')
```

We have 4 columns where distribution of values are continuous "Age",  "RestingBP"  ,"Cholestrol" , "MaxHR"

# **Pie Charts for Categorical Columns**
Pie Charts work well for column having multiple categories

```python
# autopct : auto punctuations
colors =['pink','gold','red']
df.groupby("Sex").size().plot(kind='pie',autopct='%.2f',colors=colors)
```

```python
colors =['pink','gold','red','blue']
df.groupby("ChestPainType").size().plot(kind='pie',autopct='%.2f',colors=colors)
```

```python
colors =['pink','gold','red','blue']
df.groupby("RestingECG").size().plot(kind='pie',autopct='%.2f',colors=colors)
# LVH: left ventricular hypertrophy 
```

```python
colors =['pink','gold','red','blue']
df.groupby("ST_Slope").size().plot(kind='pie',autopct='%.2f',colors=colors)
```

```python
colors =['pink','gold','red','blue']
df.groupby("HeartDisease").size().plot(kind='pie',autopct='%.2f',colors=colors)
```


# **Violin Plot for Multiple Columns Analysis**

```python
sns.violinplot(df["Age"],color='purple')
```

```python
sns.violinplot(y=df["Sex"],x=df['HeartDisease'],palette='Set2')
```

There are more number of females who are not having heart disease

```python
sns.violinplot(y=df["Age"],x=df['HeartDisease'],palette='Set3')
```

Average age of a person having heart disease is more than person not having heart disease

# **Correlation - JointPlot**

```python
sns.jointplot(x='Age',y='MaxHR',data=df)
```

```python
sns.jointplot(x='Age',y='MaxHR',data=df,kind='hex')
```

```python
# Drawing a regression (best fit line)
sns.jointplot(x='Age',y='MaxHR',data=df,kind='reg')
```

As age increases Max Heart Rate decrease

```python
# 100% correlation
sns.jointplot(x='Age',y='Age',data=df)
```

```python
# No correlation
sns.jointplot(x='HeartDisease',y='Oldpeak',data=df)
```

```python
sns.jointplot(x='Oldpeak',y='Cholesterol',data=df)
```

```python
# Line doesn't fit the data : no correlation
sns.jointplot(x='Cholesterol',y='FastingBS',kind='reg',data=df)
```

```python
sns.jointplot(x='HeartDisease',y='MaxHR',kind='reg',data=df)
```
